## Exploratory Data Analysis (EDA)

### Explore Tabel day_df

```python
day_df.describe(include="all")
```

```python
day_df.groupby(by="temp")[['casual', 'registered', 'cnt']].sum().sort_values(by="cnt", ascending=False).reset_index().head(10)
```

Tabel di atas menampilkan data jumlah penyewa sepeda (total rental sepeda) berdasarkan suhu yang teramati.

- **Deskripsi Umum**:
  - temp: Menunjukkan suhu yang dinormalisasi, dengan nilai berkisar antara 0.437500 hingga 0.710833.
  - cnt: Total jumlah rental sepeda per hari, merupakan jumlah dari **casual** (penyewa biasa) dan **registered** (penyewa terdaftar).
- **Suhu Tertinggi dan Terendah**:
    - Suhu tertinggi tercatat pada data dengan suhu **0.710833**, di mana jumlah penyewa sepeda mencapai **22.806**.
    - Suhu terendah tercatat pada suhu **0.437500**, dengan jumlah penyewa sepeda sebanyak **22.276**.

- **Hubungan Suhu dan Jumlah Penyewa**:
    - Dari data yang ditampilkan, terlihat bahwa pada suhu yang lebih tinggi, seperti **0.649167** dan **0.710833**, jumlah penyewa relatif tinggi, mencapai angka **24.216** dan **22.806**.
    - Sebaliknya, ketika suhu lebih rendah, seperti **0.514167** dan **0.437500**, jumlah penyewa sedikit lebih rendah, yaitu **22.963** dan **22.276**.

- **Pola yang Terlihat**:
    - Secara umum, terdapat kecenderungan bahwa suhu yang lebih hangat mendorong lebih banyak penyewa sepeda, meskipun ada sedikit fluktuasi.
    - Jumlah penyewa tertinggi cenderung terjadi pada suhu sedang hingga tinggi, menunjukkan bahwa kondisi cuaca yang lebih baik (hangat) mungkin lebih menarik bagi para penyewa sepeda.

Tabel di atas menampilkan data jumlah penyewa sepeda (total rental sepeda) berdasarkan suhu yang teramati pada berbagai waktu dalam sehari.

  - **hr**: Menunjukkan waktu dalam sehari (jam), dengan nilai berkisar antara **00:00** hingga **23:00**.

- **Suhu Tertinggi dan Terendah**:
    - Suhu tertinggi tercatat pada **temp** sebesar **0.710833** pada jam **18:00**, di mana jumlah penyewa sepeda mencapai **16.990**.
    - Suhu terendah tercatat pada **temp** sebesar **0.437500**, dengan jumlah penyewa sepeda sebanyak **22.276**.

- **Hubungan Suhu dan Jumlah Penyewa**:
    - Dari data yang ditampilkan, terlihat bahwa pada suhu yang lebih tinggi, seperti **0.710833** dan **0.696667**, jumlah penyewa relatif tinggi, mencapai angka **16.990** dan **16.651** pada jam **18:00** dan **19:00**.
    - Sebaliknya, pada suhu yang lebih rendah, seperti **0.514167** dan **0.437500**, jumlah penyewa sepeda sedikit lebih rendah, sekitar **22.963** dan **22.276**.

- **Pola yang Terlihat**:
    - Secara umum, jumlah penyewa sepeda lebih tinggi pada suhu hangat, terutama pada jam-jam sibuk di sore hari antara **17:00 hingga 19:00**.
    - Pola ini menunjukkan bahwa suhu yang lebih hangat dan waktu sore mendorong lebih banyak penyewa sepeda, mungkin karena aktivitas pulang kerja dan kondisi cuaca yang lebih nyaman.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah ada korelasi antara suhu dan jumlah perental sepeda?

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='temp', y='cnt', data=day_df, scatter_kws={'s':10}, line_kws={"color": "red"})
plt.title('Hubungan antara Suhu dan Jumlah Rental Sepeda')
plt.xlabel('Suhu')
plt.ylabel('Jumlah Rental Sepeda')
plt.grid(True)
plt.show()
```

### Pertanyaan 2: Apakah ada pola persewaan sepeda berdasarkan suhu pada waktu yang berbeda dalam sehari?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='hr', y='cnt', hue='temp', data=hour_df)
plt.title('Pola Persewaan Sepeda Berdasarkan Suhu dan Waktu')
plt.xlabel('Waktu Persewaan')
plt.ylabel('Jumlah Rental Sepeda')
plt.grid(True)
plt.show()
```

**Insight:**
- Pada Visualisasi pertanyaan pertama terlihat sebaran data menggunakan fungsi plot
- Menyesuaikan plot dengan pertanyaan

## Conclusion

**1. Apakah ada korelasi antara suhu dan jumlah perental sepeda?**
   
   Berdasarkan plot hubungan antara suhu dan jumlah rental sepeda, terlihat bahwa terdapat korelasi positif antara suhu dan jumlah persewaan sepeda. Saat suhu meningkat, jumlah penyewaan sepeda cenderung meningkat, menunjukkan bahwa suhu yang lebih hangat mendorong lebih banyak orang untuk menyewa sepeda. Regresi linier dengan garis merah menunjukkan tren kenaikan yang konsisten, meskipun terdapat beberapa variasi di sekitar garis tersebut, yang menunjukkan adanya faktor lain yang mungkin juga memengaruhi jumlah penyewaan. Secara keseluruhan, suhu tampaknya menjadi salah satu faktor penting yang mempengaruhi keputusan orang untuk menyewa sepeda.

**2. Apakah ada pola persewaan sepeda berdasarkan suhu pada waktu yang berbeda dalam sehari?**

  Persewaan sepeda cenderung mencapai puncaknya pada waktu tertentu, terutama pada sore hingga malam hari (sekitar pukul 17:00 hingga 19:00), yang kemungkinan terkait dengan waktu pulang kerja atau aktivitas rutin sore hari. Selain itu, suhu yang lebih tinggi (terlihat dari perbedaan warna) juga berhubungan dengan jumlah rental sepeda yang lebih besar. Persewaan sepeda pada suhu yang hangat dan pada jam sibuk (seperti sore hari) cenderung lebih tinggi, menegaskan bahwa faktor waktu dan kondisi cuaca memainkan peran penting dalam jumlah penyewaan sepeda.
